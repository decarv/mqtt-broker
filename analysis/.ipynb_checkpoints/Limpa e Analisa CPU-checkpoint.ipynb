```python
i = 1
with open('logs/100-8-clients-cpu.log') as f:
    cl100 = f.read()
```

```python
separator = '+++++++++++++++++++++++++++'
lst = cl100.split(separator)
lstlst = []
for el in lst:
    lstlst.append(el.split('\n'))

for lst in lstlst:
    for el in lst:
        if el == '':
            lst.remove('')

for lst in lstlst:
    if len(lst) == 0:
        lstlst.remove(lst)

for lst in lstlst:
    for i, el in enumerate(lst):
        lst[i] = el.split()

for i, lst in enumerate(lstlst):
    for j, el in enumerate(lst):
        lst[j] = [str(i)] + el

df_lst = []
for i, lst in enumerate(lstlst):
    df_lst.append(pd.DataFrame(lstlst[i]))

df_lst[105]

for i in range(len(df_lst)):
    df_lst[i].iloc[:,3] = df_lst[i].iloc[:,3].astype(np.float64)

clean = []

for i in range(len(df_lst)):
    clean.append([df_lst[i][3].sum(), df_lst[i][3].mean(), df_lst[i][10].max()])

df = pd.DataFrame(clean, columns=['sumCPU%', 'meanCPU%', 'uptime'])

df['uptime'] = pd.to_datetime(df['uptime'], format='%M:%S')
```

```python
sb.lineplot(data=df, x="uptime", y="meanCPU%")
plt.title(str(i) + ". 100 Subs e 1000 Pubs");
```
